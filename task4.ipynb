```python
import seaborn as sns
import pandas as pd

df = sns.load_dataset("car_crashes")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt

top_states = df.groupby("abbrev")["total"].sum().sort_values(ascending=False)

top_states.head(10).plot(kind="bar", figsize=(10,5))
plt.title("Top Accident Hotspot States")
plt.ylabel("Total Accidents")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df["alcohol"], df["total"])
plt.title("Alcohol vs Accidents")
plt.xlabel("Alcohol Influence")
plt.ylabel("Total Accidents")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df["speeding"], df["total"])
plt.title("Speeding vs Accidents")
plt.xlabel("Speeding")
plt.ylabel("Total Accidents")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Between Accident Factors")
plt.show()
```
